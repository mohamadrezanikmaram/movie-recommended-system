# **add libraries** 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

# **read clean movie dataframe**

```python
movie_df = pd.read_csv('clean_movies.csv')
movie_df
```

# **building a list of user  for implementation ssystem**

```python
import numpy as np

num_users = 10

users = [f'User_{i}' for i in range(1, num_users+1)]
movie_df['User'] = np.random.choice(users, size=len(movie_df))
```

# **show sample**

```python
movie_df['User']
```

# **random choice number of users or samples for checking and recommending**

```python
sample_df = movie_df.sample(n=20, random_state=42)
sample_df
```

# **kepping needful columns for system and delete extra columns**

```python
needed_columns = ['User', 'Series_Title', 'Genre', 'IMDB_Rating']
sample_df = sample_df[needed_columns]
sample_df
```

# **sorting rows with imbd_rate datas**

```python
sample_df = sample_df.sort_values(by='IMDB_Rating', ascending=False)
sample_df
```

# **reset indexes**

```python
sample_df = sample_df.reset_index(drop=True)
sample_df
```

# **genre list**

```python
#all genres add to a list
all_genres = sample_df['Genre'].str.split(',').explode()

#delete spaces
all_genres = all_genres.str.strip()

#get unique genre 
unique_genres = all_genres.unique()

#sorting list A--z
unique_genres = sorted(unique_genres)

print(unique_genres)
```

# **show with crosstab method**

```python
pd.crosstab(index=sample_df['User'],
                           columns=sample_df['Genre'],
                           )
```

# **founder function**

```python
def recommend_by_genre(genre, n=5):
   
    filtered = sample_df[sample_df['Genre'].str.contains(genre, case=False)]
    
    filtered = filtered.sort_values(by='IMDB_Rating', ascending=False)

    return filtered[['Series_Title', 'Genre', 'IMDB_Rating']].head(n)


recommend_by_genre('Biography, Crime, Drama', n=5)
```
